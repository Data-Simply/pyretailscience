# Heatmap Plot Gallery

The heatmap plot is used for visualizing 2D data structures as color-coded matrices with automatic cell value annotations. It's ideal for displaying relationships, patterns, and distributions across two categorical dimensions.

Heatmap plots excel at:
- **Correlation Analysis**: Visualize relationships between variables or metrics
- **Migration Matrices**: Track customer movement between segments over time  
- **Performance Grids**: Compare metrics across categories, regions, or time periods
- **Confusion Matrices**: Display classification results and model performance

```python
import matplotlib.pyplot as plt
import pandas as pd

from pyretailscience.plots import heatmap
```

## Basic Heatmap

Create a simple heatmap from a DataFrame. The DataFrame index becomes the y-axis and columns become the x-axis.

---

```python
# Create sample sales performance data across regions and quarters
# fmt: off
sales_performance = pd.DataFrame(
    {
        "Q1": [125.5, 98.3, 87.2, 134.1],
        "Q2": [142.8, 108.7, 95.4, 156.3],
        "Q3": [138.2, 112.5, 89.6, 148.9],
        "Q4": [165.7, 125.4, 102.3, 178.2],
    },
    index=["North Region", "South Region", "West Region", "East Region"],
)
# fmt: on

ax = heatmap.plot(
    sales_performance,
    cbar_label="Sales Performance ($000s)",
    title="Quarterly Sales Performance by Region",
    x_label="Quarter",
    y_label="Region",
)
plt.show()
```

## Customer Migration Matrix

Visualize customer movement between segments over time. This example shows how customers transition between different value tiers.

---

```python
# Create sample customer migration data (percentage of customers moving between segments)
# fmt: off
migration_data = pd.DataFrame(
    {
        "Bronze": [0.75, 0.15, 0.08, 0.02],
        "Silver": [0.12, 0.68, 0.18, 0.02],
        "Gold": [0.03, 0.22, 0.65, 0.10],
        "Platinum": [0.01, 0.05, 0.25, 0.69],
    },
    index=["Bronze", "Silver", "Gold", "Platinum"],
)
# fmt: on

ax = heatmap.plot(
    migration_data,
    cbar_label="Migration Probability",
    title="Customer Segment Migration Matrix (Q1 to Q2)",
    x_label="Current Quarter Segment",
    y_label="Previous Quarter Segment",
)
plt.show()
```

## Custom Colorbar Format

Customize the colorbar display format using the `cbar_format` parameter. This example shows percentage formatting.

---

```python
# Create sample conversion rate data across channels and time periods
# fmt: off
conversion_rates = pd.DataFrame(
    {
        "Week 1": [0.0245, 0.0189, 0.0312, 0.0198],
        "Week 2": [0.0278, 0.0201, 0.0334, 0.0215],
        "Week 3": [0.0256, 0.0195, 0.0298, 0.0187],
        "Week 4": [0.0289, 0.0223, 0.0356, 0.0231],
    },
    index=["Email Campaign", "Social Media", "Google Ads", "Display Ads"],
)
# fmt: on

ax = heatmap.plot(
    conversion_rates,
    cbar_label="Conversion Rate",
    cbar_format="{x:.1%}",  # Format as percentage with 1 decimal place
    title="Weekly Conversion Rates by Marketing Channel",
    x_label="Week",
    y_label="Marketing Channel",
)
plt.show()
```

## Correlation Matrix

Visualize relationships between different business metrics. The heatmap automatically handles the symmetric nature of correlation data.

---

```python
# Create sample correlation matrix for business metrics
# fmt: off
correlation_matrix = pd.DataFrame(
    {
        "Revenue": [1.00, 0.85, 0.72, 0.68, 0.45],
        "Orders": [0.85, 1.00, 0.81, 0.59, 0.52],
        "Customers": [0.72, 0.81, 1.00, 0.43, 0.38],
        "AOV": [0.68, 0.59, 0.43, 1.00, 0.31],
        "Marketing Spend": [0.45, 0.52, 0.38, 0.31, 1.00],
    },
    index=["Revenue", "Orders", "Customers", "AOV", "Marketing Spend"],
)
# fmt: on

ax = heatmap.plot(
    correlation_matrix,
    cbar_label="Correlation Coefficient",
    cbar_format="{x:.2f}",
    title="Business Metrics Correlation Matrix",
    x_label="Metric",
    y_label="Metric",
)
plt.show()
```

```python
# Create sample satisfaction data with long column names (triggers auto-rotation)
# fmt: off
satisfaction_scores = pd.DataFrame(
    {
        "Website Experience": [4.2, 3.8, 4.1, 3.9],
        "Customer Support": [4.5, 4.3, 4.6, 4.2],
        "Product Quality": [4.7, 4.4, 4.8, 4.5],
        "Delivery Service": [3.9, 4.1, 3.8, 4.0],
        "Return Process": [3.6, 3.9, 3.5, 3.8],
    },
    index=["Q1 2024", "Q2 2024", "Q3 2024", "Q4 2024"],
)
# fmt: on

ax = heatmap.plot(
    satisfaction_scores,
    cbar_label="Satisfaction Score",
    cbar_format="{x:.1f}",
    title="Customer Satisfaction Scores by Service Area",
    x_label="Service Area",
    y_label="Quarter",
    figsize=(8, 4),
)
plt.show()
```

```python
# Create sample inventory turnover data across product lines and seasons
# fmt: off
inventory_turnover = pd.DataFrame(
    {
        "Spring": [2.8, 4.1, 3.2, 5.6, 3.9],
        "Summer": [3.4, 6.8, 4.1, 7.2, 4.5],
        "Fall": [4.2, 5.3, 3.8, 6.1, 4.7],
        "Winter": [3.1, 3.9, 4.4, 4.8, 5.2],
    },
    index=["Fashion", "Electronics", "Home Decor", "Sporting Goods", "Books"],
)
# fmt: on

ax = heatmap.plot(
    inventory_turnover,
    cbar_label="Turnover Rate",
    cbar_format="{x:.1f}x",  # Custom format with 'x' suffix
    title="Seasonal Inventory Turnover by Product Line",
    x_label="Season",
    y_label="Product Line",
    source_text="Source: Inventory Management System - FY 2024",
)
plt.show()
```
